Реализовать возможность «отката» ходов. С помощью специальной команды можно возвращаться на ход (или заданное количество ходов) назад вплоть до начала партии.

```python
rows = dict(zip('87654321', range(8))) # словарь для определения позиции на игровом поле (строки)
columns = dict(zip('ABCDEFGH', range(8))) # словарь для определения позиции на игровом поле (столбца)
name = ['пешка','ладья','конь','слон','ферзь','король']
white_pieces = dict(zip('PRNBQK', [i for i in name]))
black_pieces = dict(zip('prnbqk', [i for i in name]))
moves = [] # список, куда будут записываться ходы

# функция для вывода игрового поля
def prnt(): 
    for i,j in dict(zip(range(1,9), field)).items():
        playing_field[i][1:9] = j
    for i in playing_field:
        print(' '.join(i))

# функция для проверки, чтоб фигура не могла перепрыгивать через другие, двигаясь прямо вверх/вниз или вправо/влево
def examination1(f0, f1):
    start = (rows[f0[1]], columns[f0[0]])
    end = (rows[f1[1]], columns[f1[0]])    
    start_row, start_col = start
    end_row, end_col = end
    if start_row == end_row: # A1 -> H1
        col_step = 1 if end_col > start_col else -1
        current_col = start_col + col_step
        while current_col != end_col:
            if field[start_row][current_col] != '.':
                print('Такой ход невозможен!')
                return False
            current_col += col_step
    elif start_col == end_col: # A1 -> A5
        row_step = 1 if end_row > start_row else -1
        current_row = start_row + row_step
        while current_row != end_row:
            if field[current_row][start_col] != '.':
                print('Такой ход невозможен!')
                return False
            current_row += row_step            
    return False

# функция для проверки, чтоб фигура не могла перепрыгивать через другие, двигаясь по диагонали вверх/вниз или вправо/влево
def examination2(f0, f1):
    start = (rows[f0[1]], columns[f0[0]])
    end = (rows[f1[1]], columns[f1[0]])    
    start_row, start_col = start
    end_row, end_col = end
    # Проверяем, что ход по диагонали
    if abs(start_row - end_row) != abs(start_col - end_col):
        print('Такой ход невозможен!')
        return False

    # Проверяем, что путь не заблокирован другими фигурами
    # определяем направление движения по строкам и столбцам: если конечная позиция находится ниже или 
    # правее начальной позиции, то row_step и col_step равны 1, иначе -1
    row_step = 1 if end_row > start_row else -1
    col_step = 1 if end_col > start_col else -1
    # значения, соответствующие соседней клетке по направлению движения от начальной позици
    current_row, current_col = start_row + row_step, start_col + col_step
    while current_row != end_row and current_col != end_col:
        if field[current_row][current_col] != '.':
            print('Такой ход невозможен!')
            return False
        current_row += row_step
        current_col += col_step
    return False    
    
# функция для осуществления хода
def move(c): 
    while True:
        try:
            f0 = input('Введите позицию выбранной фигуры: ') # позиция фигуры (E2)
            f1 = input('Введите целевую позицию для выбранной фигуры: ') # целевая позиция фигуры (E3)
            moves.append((f0,f1)) # записываем ход 
            # проверка правильности хода
            
            if field[rows[f0[1]]][columns[f0[0]]] == 'P': # БЕЛАЯ ПЕШКА
                if not(int(f0[1]) == int(f1[1]) - 1 and f0[0] == f1[0]): 
                # E2 -> E3
                    print('Невозможный ход для белой пешки!')
                    continue
                else:
                    break
            elif field[rows[f0[1]]][columns[f0[0]]] == 'p': # ЧЕРНАЯ ПЕШКА
                if not(int(f0[1]) == int(f1[1]) + 1 and f0[0] == f1[0]):   
                # E7 -> E6
                    print('Невозможный ход для черной пешки!')
                    continue
                else:
                    break
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'R' or field[rows[f0[1]]][columns[f0[0]]] == 'r': # ЛАДЬЯ
                if examination1(f0, f1):
                    break
                else:
                    continue
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'N' or field[rows[f0[1]]][columns[f0[0]]] == 'n': # КОНЬ
                if not((abs(int(f0[1])-int(f1[1]))==2 and abs(columns[f0[0]]-columns[f1[0]])==1) or (abs(int(f0[1])-int(f1[1]))==1 and abs(columns[f0[0]]-columns[f1[0]])==2)): 
                # E4 -> F6
                # E4 -> G5
                    print('Невозможный ход для коня!')
                    continue
                else:
                    break
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'B' or field[rows[f0[1]]][columns[f0[0]]] == 'b': # СЛОН
                if examination2(f0, f1):
                    break
                else:
                    continue
                
            elif field[rows[f0[1]]][columns[f0[0]]] == 'Q' or field[rows[f0[1]]][columns[f0[0]]] == 'q': # ФЕРЗЬ
                # сочетание ладьи и слона
                # A1 -> A5
                # A1 -> H1
                # E4 -> C2 (44 -> 26: 4-2 = -(4-6))
                # E4 -> C6 (44 -> 22: 4-2 = 4-2)
                if examination1(f0, f1) or examination2(f0, f1):
                    break
                else:
                    continue
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'K' or field[rows[f0[1]]][columns[f0[0]]] == 'k': # КОРОЛЬ
                if not((f0[0]==f1[0] and abs(int(f0[1])-int(f1[1]))==1) or (f0[1]==f1[1] and abs(columns[f0[0]]-columns[f1[0]])==1) or (abs(int(f0[1])-int(f1[1]))==1 and abs(columns[f0[0]]-columns[f1[0]])==1)): 
                # E4 -> E3, E3 -> E4
                # E4 -> F4, F4 -> E4
                # E4 -> F5, F5 -> E4
                    print('Невозможный ход для короля!')
                    continue
                else:
                    break
                    
        except IndexError:
            print('Ход невозможен!')
            continue
        except KeyError:
            print('Неверно введена позиция!')
            continue
    # сообщение о потере фигуры
    if field[rows[f1[1]]][columns[f1[0]]] != '.': 
        if c%2 == 0:
            print(f'Черная фигура {black_pieces[field[rows[f1[1]]][columns[f1[0]]]]} съедена!')
        else:
            print(f'Белая фигура {white_pieces[field[rows[f1[1]]][columns[f1[0]]]]} съедена!')
    # изменяем поле
    field[rows[f1[1]]][columns[f1[0]]] = field[rows[f0[1]]][columns[f0[0]]] # переставляем фигуру
    field[rows[f0[1]]][columns[f0[0]]] = '.' # "заметаем следы"
        
# функция для "отката" ходов .........
def reversing(c):
    while True:
        try:
            n = int(input('Введите количество ходов для отката: ')) # n - количество откатов ходов
            for i in range(n):
                try:
                    field[rows[moves[-1][0][1]]][columns[moves[-1][0][0]]] = field[rows[moves[-1][1][1]]][columns[moves[-1][1][0]]] # переставляем фигуру
                    field[rows[moves[-1][1][1]]][columns[moves[-1][1][0]]] = '.' # "заметаем следы"
                    # moves[-1][0] - f0
                    # moves[-1][1] - f1
                    c[0] -= 1 # возвращаемся на ход назад
                    print(f'Отменен ход {moves.pop()}') # удаляем шаг, который был отменен
                    prnt()
                except IndexError:
                    print('Откат не возможен из-за отсутствия предшествующих ходов!')
        except ValueError:
            print('Неверно введено количество ходов для отката!')
            continue
        break
        

# функция, проверяющая продолжение партии
def the_game_continues(c):
    a = input('''Нажмите Enter для продолжения игры, напечатайте Stop для окончания партии 
    или Reverse для "отката" хода: ''')
    if a == 'Stop':
        print(f'Партия окончена. Всего ходов было {c[0]}')
        return False
    if a == 'Reverse':
        reversing(c)
        return True
    for row in field:
        if any(white_piece in row for white_piece in white_pieces):
            return True
        elif any(black_piece in row for black_piece in black_pieces):
            return True
```

```python
def Сhess(): 
    prnt() # выводим начальное игровое поле
    c = [0] # переменная для подсчёта ходов (ввиде списка, чтоб ее значение могла изменять вложенная функция)
    while the_game_continues(c):  
        if c[0]%2 == 0:
            print(f'    Ход №{c[0]+1}')
            print('ХОДЯТ БЕЛЫЕ')
            move(c[0])
            prnt()
        else:
            print(f'    Ход №{c[0]+1}')
            print('ХОДЯТ ЧЕРНЫЕ')
            move(c[0])
            prnt()
        c[0] += 1
```
