```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axisartist.axislines import AxesZero
```

```python
figure = plt.figure(figsize = (5, 5), dpi = 800)
axs = figure.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    # adds X and Y-axis from the origin
    axs.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    axs.axis[direction].set_visible(False)

axs.set_xticks([])
axs.set_yticks([])
axs.set_xlim([-10, 10])
axs.set_ylim([-10, 10])

#1------------
def generate_rectangle(start, end):
    x = [i for i in range(start, end, 1)]
    y = [0] * (abs(start) + abs(end))
    axs.scatter(x, y, marker = 's', color = 'blue')
    
def generate_triangle(start, end):
    x = [i for i in range(start, end, 1)]
    y = [0] * (abs(start) + abs(end))
    axs.scatter(x, y, marker = '^', color = 'red')

def generate_hexagon(start, end):
    x = [i for i in range(start, end, 1)]
    y = [0] * (abs(start) + abs(end))
    axs.scatter(x, y, marker = 'H', color = 'green')


#2------------
def tr_translate(count, start, end, type_figure):
    type_figure = str(type_figure)
    if type_figure == 'треугольник':
        x = [i for i in range(start, end, 1)]
        y = [count]*(abs(start) + abs(end))
        axs.scatter(x, y, marker = '^', color = 'red')
            
    elif type_figure == 'квадрат':
        x = [i for i in range(start, end, 1)]
        y = [count]*(abs(start) + abs(end))
        axs.scatter(x, y, marker = 's', color = 'blue')
            
    elif type_figure == 'шестиугольник':
        x = [i for i in range(start, end, 1)]
        y = [count]*(abs(start) + abs(end))
        axs.scatter(x, y, marker = 'H', color = 'green')

def tr_rotate(start, end, type_figure):
    type_figure = str(type_figure)
    if type_figure == 'треугольник':
        x = [0] * (abs(start) + abs(end))
        y = [i for i in range(start, end, 1)]
        axs.scatter(x, y, marker = '^', color = 'red')
            
    elif type_figure == 'квадрат':
        x = [0] * (abs(start) + abs(end))
        y = [i for i in range(start, end, 1)]
        axs.scatter(x, y, marker = 's', color = 'blue')
            
    elif type_figure == 'шестиугольник':
        x = [0] * (abs(start) + abs(end))
        y = [i for i in range(start, end, 1)]
        axs.scatter(x, y, marker = 'H', color = 'green')

def tr_symmetry(ox, start, end, type_figure):
    type_figure = str(type_figure)
    if type_figure == 'треугольник':
        x = [ox] * (abs(start) + abs(end))
        y = [i for i in range(start, end, 1)]
        x_sym = [-ox] * (abs(start) + abs(end))
        axs.scatter(x, y, marker = '^', color = 'red')
        axs.scatter(x_sym, y, marker = '^', color = 'red')
            
    elif type_figure == 'квадрат':
        x = [ox] * (abs(start) + abs(end))
        y = [i for i in range(start, end, 1)]
        x_sym = [-ox] * (abs(start) + abs(end))
        axs.scatter(x, y, marker = 's', color = 'blue')
        axs.scatter(x_sym, y, marker = 's', color = 'blue')
            
    elif type_figure == 'шестиугольник':
        x = [ox] * (abs(start) + abs(end))
        y = [i for i in range(start, end, 1)]
        x_sym = [-ox] * (abs(start) + abs(end))
        axs.scatter(x, y, marker = 'H', color = 'green')
        axs.scatter(x_sym, y, marker = 'H', color = 'green')

    
#3------------
def three_paral(start, end):
    x = [i for i in range(start, end, 1)]
    for y in range(3):
        axes = [(y - 4) + i for i in range(abs(start) + abs(end))]
        axs.scatter(x, axes, marker = 's', color = 'blue')

def cross(k, start, end):
    x = [i for i in range(start, end, 1)]
    for y in range(2):
        if y % 2 == 0:
            axes = [(0 + i)*k for i in range(abs(start) + abs(end))]
            axs.scatter(x, axes, marker = 's', color = 'blue')
        else:
            axes = [(0 + i)*k for i in range(abs(start) + abs(end), 0, -1)]
            axs.scatter(x, axes, marker = 's', color = 'blue')

def triangle_paral(start, end):
    x = [i for i in range(start, end, 1)]
    for y in range(2):
        if y % 2 == 0:
            axes = [y] * (abs(start) + abs(end))
            axs.scatter(x, axes, marker = '^', color = 'red')
        else:
            axes = [y] * (abs(start) + abs(end))
            axs.scatter(x, axes, marker = 'v', color = 'red')

#1 figures
generate_rectangle(-3, 7)
# generate_triangle(-3, 9)
# generate_hexagon(-2, 2)
#2 translates
# tr_translate(5, -3, 9, 'треугольник')
# tr_rotate(-3, 7, 'квадрат')
tr_symmetry(4, -3, 7, 'шестиугольник')
#3 paint
# triangle_paral(-3, 7)
cross(0.5, -3, 5)
# three_paral(-3, 4)
```
