Реализовать программу, которая позволяет играть в шахматы на компьютере. Взаимодействие с программой производится через консоль (базовый вариант). Игровое поле изображается в виде 8 текстовых строк, плюс строки с буквенным обозначением столбцов (см. пример на Рис. 1) и перерисовывается при каждом изменении состояния поля. При запросе данных от пользователя программа сообщает, что ожидает от пользователя (например, позицию фигуры для следующего хода белыми; целевую позицию выбранной фигуры) и проверяет корректность ввода (допускаются только ходы соответствующие правилам шахмат; поддержка рокировки, сложных правил для пешек и проверки мата вынесена в отдельные пункты). Программа должна считать количество сделанных ходов. 

Сама программа НЕ ходит: т.е. не пытается выполнить ходы за одну из сторон, а предоставляет поочередно вводить ходы за белых и черных.

#####  A B C D E F G H
##### 8 r n b q k b n r 8 черные
##### 7 p p p p p p p p 7
##### 6 . . . . . . . . 6
##### 5 . . . . . . . . 5
##### 4 . . . . . . . . 4
##### 3 . . . . . . . . 3
##### 2 P P P P P P P P 2
##### 1 R N B Q K B N R 1 белые
#####  A B C D E F G H

R,r - ладья
N,n - конь
B,b - слон
Q,q - ферзь
K,k - король
P,p - пешка

```python
playing_field = [
[' ','A','B','C','D','E','F','G','H',' '],
['8','r','n','b','q','k','b','n','r','8'],
['7','p','p','p','p','p','p','p','p','7'],
['6','.','.','.','.','.','.','.','.','6'],
['5','.','.','.','.','.','.','.','.','5'],
['4','.','.','.','.','.','.','.','.','4'],
['3','.','.','.','.','.','.','.','.','3'],
['2','P','P','P','P','P','P','P','P','2'],
['1','R','N','B','Q','K','B','N','R','1'],
[' ','A','B','C','D','E','F','G','H',' ']
]
field = [i[1:9] for i in playing_field[1:9]]
```

```python
rows = dict(zip('87654321', range(8))) # строки игрового поля
columns = dict(zip('ABCDEFGH', range(8))) # столбцы игрового поля
name = ['пешка','ладья','конь','слон','ферзь','король']
white_pieces = dict(zip('PRNBQK', [i for i in name]))
black_pieces = dict(zip('prnbqk', [i for i in name]))


def prnt(): 
    '''вывод игрового поля'''
    for i,j in dict(zip(range(1,9), field)).items():
        playing_field[i][1:9] = j
    for i in playing_field:
        print(' '.join(i))

```

```python
def correction_1(f0, f1):
    '''проверка корректности хода вверх\вниз\влево\вправо'''
    start = (rows[f0[1]], columns[f0[0]])
    end = (rows[f1[1]], columns[f1[0]])    
    start_row, start_col = start
    end_row, end_col = end
    if start_row == end_row: # A1 -> H1
        col_step = 1 if end_col > start_col else -1
        current_col = start_col + col_step
        while current_col != end_col:
            if field[start_row][current_col] != '.':
                print('impossible move!')
                return False
            current_col += col_step
    elif start_col == end_col: # A1 -> A5
        row_step = 1 if end_row > start_row else -1
        current_row = start_row + row_step
        while current_row != end_row:
            if field[current_row][start_col] != '.':
                print('impossible move!')
                return False
            current_row += row_step            
    return False 

```

```python
def correction_2(f0, f1):
    '''проверка корректности хода по диагонали'''
    start = (rows[f0[1]], columns[f0[0]])
    end = (rows[f1[1]], columns[f1[0]])    
    start_row, start_col = start
    end_row, end_col = end
    # Проверяем, что ход по диагонали
    if abs(start_row - end_row) != abs(start_col - end_col):
        print('impossible move!')
        return False

    '''проверка блока фигур'''
    row_step = 1 if end_row > start_row else -1
    col_step = 1 if end_col > start_col else -1
    # значения, соответствующие соседней клетке по направлению движения от начальной позици
    current_row, current_col = start_row + row_step, start_col + col_step
    while current_row != end_row and current_col != end_col:
        if field[current_row][current_col] != '.':
            print('impossible move!')
            return False
        current_row += row_step
        current_col += col_step
    return False
        
```

```python
def move(c): 
    '''реализация хода'''
    
    while True:
        try:
            f0 = input('input start piece position: ') # позиция фигуры (E2)
            f1 = input('input end piece position: ') # целевая позиция фигуры (E3)
            # проверка правильности хода
            
            if field[rows[f0[1]]][columns[f0[0]]] == 'P': # БЕЛАЯ ПЕШКА
                if not(int(f0[1]) == int(f1[1]) - 1 and f0[0] == f1[0]): 
                # E2 -> E3
                    print('impossible white pawn move!')
                    continue
                else:
                    break
            elif field[rows[f0[1]]][columns[f0[0]]] == 'p': # ЧЕРНАЯ ПЕШКА
                if not(int(f0[1]) == int(f1[1]) + 1 and f0[0] == f1[0]):   
                # E7 -> E6
                    print('impossible black pawn move!')
                    continue
                else:
                    break
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'R' or field[rows[f0[1]]][columns[f0[0]]] == 'r': # ЛАДЬЯ
                if correction_1(f0, f1):
                    break
                else:
                    continue
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'N' or field[rows[f0[1]]][columns[f0[0]]] == 'n': # КОНЬ
                if not((abs(int(f0[1]) - int(f1[1])) == 2 and abs(columns[f0[0]] - columns[f1[0]]) == 1) or (abs(int(f0[1]) - int(f1[1])) == 1 and abs(columns[f0[0]] - columns[f1[0]]) == 2)): 
                # E4 -> F6
                # E4 -> G5
                    print('impossible knight move!')
                    continue
                else:
                    break
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'B' or field[rows[f0[1]]][columns[f0[0]]] == 'b': # СЛОН
                if correction_2(f0, f1):
                    break
                else:
                    continue
                
            elif field[rows[f0[1]]][columns[f0[0]]] == 'Q' or field[rows[f0[1]]][columns[f0[0]]] == 'q': # ФЕРЗЬ
                # сочетание ладьи и слона
                # A1 -> A5
                # A1 -> H1
                # E4 -> C2 (44 -> 26: 4-2 = -(4-6))
                # E4 -> C6 (44 -> 22: 4-2 = 4-2)
                if correction_1(f0, f1) or correction_2(f0, f1):
                    break
                else:
                    continue
                    
            elif field[rows[f0[1]]][columns[f0[0]]] == 'K' or field[rows[f0[1]]][columns[f0[0]]] == 'k': # КОРОЛЬ
                if not((f0[0] == f1[0] and abs(int(f0[1]) - int(f1[1])) == 1) or (f0[1] == f1[1] and abs(columns[f0[0]] - columns[f1[0]]) == 1) or (abs(int(f0[1]) - int(f1[1])) == 1 and abs(columns[f0[0]] - columns[f1[0]]) == 1)): 
                # E4 -> E3, E3 -> E4
                # E4 -> F4, F4 -> E4
                # E4 -> F5, F5 -> E4
                    print('impossible king move!')
                    continue
                else:
                    break
                    
        except IndexError:
            print('impossible move!')
            continue
        except KeyError:
            print('wrong position input!')
            continue
    # сообщение о потере фигуры
    if field[rows[f1[1]]][columns[f1[0]]] != '.': 
        if c%2 == 0:
            print(f'Black piece {black_pieces[field[rows[f1[1]]][columns[f1[0]]]]} has eatend!')
        else:
            print(f'Black piece {white_pieces[field[rows[f1[1]]][columns[f1[0]]]]} has eaten!')
    # изменяем поле
    field[rows[f1[1]]][columns[f1[0]]] = field[rows[f0[1]]][columns[f0[0]]] # переставляем фигуру
    field[rows[f0[1]]][columns[f0[0]]] = '.' # "заметаем следы"


```

```python
def the_game_continues(c):
    '''проверка на остановку партии'''
    
    a = input('Press Enter to continue the game or type Stop to end the game:')
    if a == 'Stop':
        print(f'Game is over\nMoves overall: {c}')
        return False
    for row in field:
        if any(white_piece in row for white_piece in white_pieces):
            return True
        elif any(black_piece in row for black_piece in black_pieces):
            return True
```

```python
def Сhess(): 
    prnt() # рисуем поле
    c = 0 # счетчик ходов
    while the_game_continues(c):  
        if c%2 == 0:
            print(f'    Move №{c+1}')
            print("White's move")
            move(c)
            prnt()
        else:
            print(f'    Move №{c+1}')
            print("Black's move")
            move(c)
            prnt()
        c += 1
```
